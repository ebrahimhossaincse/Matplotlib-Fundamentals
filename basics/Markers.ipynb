## Point Marker (.)

```python
import matplotlib.pyplot as plt

# Sample data
x = [22, 1, 7, 2, 21, 11, 14, 5]
y = [24, 2, 12, 5, 5, 5, 9, 12]

# Plot with Point Marker
plt.scatter(x, y, marker='.', s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Point Marker (.)')
plt.show()
```
